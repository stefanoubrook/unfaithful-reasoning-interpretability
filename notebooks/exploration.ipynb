# Exploration

Notebook for exploring model outputs and activations.

```python
import json
import numpy as np
```

```python
# Load scored outputs
records = []
with open("../data/processed/scored_outputs.jsonl") as f:
    for line in f:
        records.append(json.loads(line))

print(f"Total examples: {len(records)}")
print(f"Correct: {sum(1 for r in records if r['correct'])}")
```

```python
# Load activations
data = np.load("../results/activations.npz")
activations = data["activations"]
labels = data["labels"]
print(f"Activations shape: {activations.shape}")
print(f"Labels shape: {labels.shape}")
```
